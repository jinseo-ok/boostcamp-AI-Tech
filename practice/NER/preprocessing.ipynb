```python
import json
```

```python
with open("./data/raw/SXNE1902007240.json","r") as f1,open("./data/raw/NXNE1902008030.json","r") as f2:
    data1=json.load(f1)
    data2=json.load(f2)
    
datas=[data1,data2]
```

```python
results=[]
for data in datas:
    for doc in data['document']:
        ex_text=doc['sentence']
        
        for txt_dict in ex_text:
            labels=[]
            labels_=[]
            for j in txt_dict["form"]:
                if j== " ":
                    labels.append("B")
                    continue
                labels.append("O")
            for k in txt_dict["NE"]:
                data=list(k.values())
                BI_indicator=0 
                if data[2] in ['DT','LC','OG','PS','QT','TI','O']:
                  for p in range(data[3],data[4]):
                      if BI_indicator==0:
                          if labels[p]=="B":
                              BI_indicator=1
                          else:
                              labels[p]=data[2]+"-B"
                      else:
                          if not labels[p]=="B":
                              labels[p]=data[2]+"-I"
            for i in labels:
                if len(labels_)==0:
                    labels_.append(i)
                else:
                    if i=="B":
                        labels_.append(i)
                    if labels_[-1]==i:
                        continue
                    if labels_[-1]=="B":
                        labels_.append(i)
            while "B" in labels_:
                labels_.remove("B")
            results.append((txt_dict["form"],labels_))
```

```python
train=results[:int(len(results)*0.9)]
dev=results[int(len(results)*0.9):]
```

```python
with open("./data/custom-ner/train.tsv",'w') as f:
    for i in train:
        if len(i[1])<1:
            continue
        f.write(i[0])
        f.write("\t")
        f.write(' '.join(i[1]))
        f.write("\n")
```

```python
with open("./data/custom-ner/dev.tsv",'w') as f:
    for i in dev:
        if len(i[1])<1:
            continue
        f.write(i[0])
        f.write("\t")
        f.write(' '.join(i[1]))
        f.write("\n")
```
