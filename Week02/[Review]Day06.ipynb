# 1강 퀴즈

```python
# 1. 다음 벡터의 L1-노름을 구하시오
```

```python
x = np.array([-1, 2, -3])
```

```python
plt.figure(figsize = (10, 10))
ax = plt.axes(projection='3d')
ax.scatter(x[0], x[1], x[2], color = 'blue', marker='o', s=15, cmap='Greens')
ax.scatter(0, 0, 0, color = 'red', marker='o', s=15, cmap='Greens')
# plt.plot(x,[0, 0, 0])
plt.show()
```

```python
def l1_norm(x):
    x_norm = np.abs(x)
    x_norm = np.sum(x_norm)
    return x_norm

l1_norm(x)
```

```python
# 2. 다음 벡터의 L2-노름을 구하시오
```

```python
plt.scatter(-6, -8)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(y = 0, color = 'black', linewidth = 1)
plt.axvline(x = 0, color = 'black', linewidth = 1)
plt.plot([0,-6], [0, -8])
plt.show()
```

```python
def l2_norm(x):
    x_norm = x * x
    x_norm = np.sum(x_norm)
    x_norm = np.sqrt(x_norm)
    return x_norm
```

```python
x = np.array([-6, -8])
l2_norm(x)
```

```python
# 3. L2-노름을 이용하여 다음 두 벡터 사이의 거리를 구하시오 (정수값으로 입력)
```

```python
plt.scatter([-3, 0], [0, 4])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(y = 0, color = 'black', linewidth = 1)
plt.axvline(x = 0, color = 'black', linewidth = 1)
plt.plot([0,-3], [0, 0])
plt.plot([0, 0], [0, 4])

plt.show()
```

```python
plt.scatter([-3, 0], [0, 4])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(y = 0, color = 'black', linewidth = 1)
plt.axvline(x = 0, color = 'black', linewidth = 1)
plt.plot([0,-3], [0, 0])
plt.plot([0, 0], [0, 4])
plt.plot([-3, 0], [0, 4])
plt.show()
```

```python
x = np.array([-3, 0])
y = np.array([0, 4])
z = np.array([l2_norm(x), l2_norm(y)])
```

```python
z
```

```python
l2_norm(z)
```

```python
# 두 벡터 사이의 각도
def angle(x, y):
    v = np.inner(x, y) / (l2_norm(x) * l2_norm(y)) # 코사인 theta
    print(v)
    theta = np.arccos(v)
    return theta

angle(x, y)
```

# 응용2: 선형회귀분석

```python
choice = np.arange(-10, 10, 0.1)
```

```python
xs = np.random.choice(choice, 30)
ys = np.random.choice(choice, 30)
```

```python
plt.scatter(xs, ys)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(y = 0, color = 'black', linewidth = 1)
plt.axvline(x = 0, color = 'black', linewidth = 1)
```

```python
X = xs.reshape(-1, 1)
y = ys.reshape(-1, 1)
```

```python
np.dot(np.linalg.pinv(X), y)
```

```python
moore = X * 0.09357156
```

```python
moore
```

```python
# RMSE
np.mean(np.sqrt((moore - y)**2))
```

```python
plt.scatter(xs, ys)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(y = 0, color = 'black', linewidth = 1)
plt.axvline(x = 0, color = 'black', linewidth = 1)

plt.plot([-10, 10], [-0.9357156, 0.9357156], color = 'red')
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()

LR.fit(X, y)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
pred = LR.predict(X)
```

```python
np.mean(np.sqrt((pred - y)**2))
```
